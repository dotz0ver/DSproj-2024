```python
from sklearn import tree

# X[0,0] -> Y[0], X[1,1] -> Y[1]
X = [[0, 0], [1, 1]]
# 정답 값
Y = [0, 1]
```

```python
# DecisionTreeClassifier 모델 객체 선언 후, 설정 가능한 parameter들을 확인
clf = tree.DecisionTreeClassifier()
clf
```

```python
# X, Y에 대한 데이터 학습
clf = clf.fit(X, Y)
clf
```

```python
# predict 함수는 정답을 [0,1] 중 하나로만 나타냄
clf.predict([[2., 2.]])
```

```python
# predict_proba는 값을 비율로 출력
clf.predict_proba([[2., 2.]])
```

```python
# Decision tree 알고리즘은 숫자만 이해할 수 있으므로 0,1,2로 카테고리 분류
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()

# 시각화
X, y = iris.data, iris.target

# (스무 고개 마냥) 조건에 따라 True, False 값으로 나뉘게 되며 트리가 깊어질수록 샘플의 수가 적어지면서
# 해당 클래스가 어떤 클래스인지 분류하게 되는 것이 Decision Tree 알고리즘.
# DecisionTreeClassifier 모델을 선언 후, fit 함수 사용 시 학습 진행
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 10))
tree.plot_tree(clf, filled=True)
```
